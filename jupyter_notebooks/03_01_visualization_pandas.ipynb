```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('dark_background')
```

```python
#Daten wie im vorigen Beispiel

df1 = pd.DataFrame({"years" : [1950, 1960, 1970, 1980, 1990, 2000, 2010],
                    "gdp" : [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]})

#Standard-Diagrammtyp ist ein Linien-Diagramm
df1.plot()

#Mmh, nicht wie gewünscht - woran liegt das?
```

```python
#bauen wir uns schnell welche: Grösse und Gewicht
import random
mu = 170 #mean
sigma = 6 #stddev
sample = 100
random.seed(0)
height = [random.gauss(mu, sigma) for _ in range(sample)]  # Normalverteilung um 170cm
weight = [(h-100) * (0.75 + random.random()/2) for h in height]
# Gewicht streut gleichverteilt zwischen 3/4 und 5/4 der Grösse - 100

df2 = pd.DataFrame({"height":height, "weight":weight})
df2.plot.scatter(x="height", y="weight")
```
